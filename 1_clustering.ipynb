```python
# Importing Modules
from sklearn import datasets
import matplotlib.pyplot as plt

# Loading dataset
iris_df = datasets.load_iris()

# Features
print("Features: " , iris_df.feature_names)

# Targets
print("Target: " , iris_df.target)

# Target Names
print("Target names: " , iris_df.target_names)


# Dataset Slicing
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 2]  # Sepal Width

# Plotting
plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.show()
```

```python
# Importing Modules
from sklearn import datasets
from sklearn.cluster import KMeans

# Loading dataset
# iris_df = datasets.load_iris()

# Declaring Model
model = KMeans(n_clusters=3)

# Fitting Model
model.fit(iris_df.data)

temp_input = [[7.2, 3.5, 0.8, 1.6]]

# Predicitng a single input
predicted_label = model.predict(temp_input)

# Prediction on the entire data
all_predictions = model.predict(iris_df.data)

# Printing Predictions
print("Predicted abel:", temp_input , iris_df.target_names[predicted_label])
print("All predictions:", iris_df.target_names[all_predictions])
```

```python
# Dataset Slicing
x_axis = iris_df.data[:, 0]  # Sepal Length
y_axis = iris_df.data[:, 2]  # Sepal Width

# Plotting
plt.scatter(x_axis, y_axis, c=all_predictions)
plt.title("K-Means Clustering")
plt.show()

plt.scatter(x_axis, y_axis, c=iris_df.target)
plt.title("Original Dataset")
plt.show()
```

```python
# Importing Modules
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import pandas as pd

mergings = linkage(iris_df.data, method='complete')
```

```python
dendrogram(mergings,
           labels=iris_df.target,
           leaf_rotation=90,
           leaf_font_size=10,
           )
plt.rcParams['figure.figsize'] = [40, 20]
plt.show()
```
